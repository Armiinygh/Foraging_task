```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os 


file_path = './data/choiceData_experiment_1.csv'

df = pd.read_csv(file_path)
```

```python
labels = ['decision', 'reward', 'travelTask' , 'subject_id']

for index, label in enumerate(labels):
    
    uniqueness = df[label].nunique()
    print(f"Number of Unique values of {label}: {uniqueness}")
    print("*" *10)
    
    if index %2 == 0:
        uniqueness = df[label].unique()
        print(f"Unique values are {label}: {uniqueness}") 
        print("*" *10)
        
    if index %2 == 1:
        print(f"Max Value of the {label} is: {df[label].max()}")
        print(f"Min Value of the {label} is: {df[label].min()}") 
        print("*" *10)
        
         
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df['reward'], fill=True, color='green')
plt.title('Distribution of the rewards Column (KDE Plot)')
plt.xlabel('Wert')
plt.ylabel('Dichte')
plt.show()

```
